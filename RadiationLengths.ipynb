```python
def rads_plot(tag='epic_ip6_extended', mineta=-4.3, maxeta=-3.5, log=False, add_mat=None):



    from matplotlib import ticker as ticker
    phi=np.linspace(-np.pi, np.pi, len(nils[(tag, -4.0)]))
    eta=np.linspace(mineta, maxeta, int((maxeta-mineta)/.05+2))
    x=phi*180/np.pi
    y=eta
    z=z= [rads[(tag,int(i*20)/20)] for i in y]
    if add_mat is not None:
        z=[[z[i][j]+add_mat(eta[i],phi[j]) for j in range(len(z[0]))] for i in range(len(z))]
    
    #locator=ticker.LogLocator()
    #locator.view_limits(0.01, 25)
    contour = plt.contourf(x,y,z, 20, locator=locator if log else None)
    cb = plt.colorbar(contour)
    cb.set_label("# of $X_0$s traversed")
    plt.xlabel("$\phi$ [deg]")
    plt.ylabel("$\eta$")
    plt.xlim(-180, 180)
```

```python
#additional material not included in the dd4hep file
def add_mat(eta, phi):
    theta=get_theta(eta)
    X0s=0
    #circular flange, modeled as a solid cylinder
    z=-120
    r=np.tan(theta)*z
    
    rad_length_Al=8.897
    if r<7.612 and r>3.5:
        length=3.9624
        X0s+=length/rad_length_Al
        
    #now for the cam-shaped flange
    z=-269.72
    OR1=11.106/2
    IR1=8.2346/2
    x1=0
    
    
    OR2=2.8440
    IR2=3.1475/2
    x2=-0.025*z
    
    x=z*np.tan(theta)*np.cos(phi)
    y=z*np.tan(theta)*np.sin(phi)
    
    r1=np.hypot(x, y)
    r2=np.hypot(x-x2, y)
    
    if r1<OR1 and r1>IR1 or r2<OR2 and r2>IR2:
        length=2.54
        X0s+=length/rad_length_Al
    
    
    
    return X0s
```

```python
def rads_plot_1d(tag='beampipe_and_flange_only', mineta=-4.7, maxeta=-3.3, plot_type="nrad",**arg):
    eta=np.linspace(mineta, maxeta, int((maxeta-mineta)/.05+2))
    

    x=eta
    y= np.array([rads[(tag,int(i*20)/20)][-1] for i in eta])
    
    if min(y)<0.01:
        y-=min(y)
    
    print(y)
    
    if plot_type=="nrad":
        plt.plot(x,y, **arg)
        plt.ylabel("# of $X_0$s traversed (circular section)")
    elif plot_type=="prob_no_shower":
        plt.plot(x,np.exp(-y), **arg)
        plt.ylabel("prob. of not showering (circ. section)")
    
    
    plt.xlabel("$\eta$")
    #plt.ylim(0)
    plt.legend()
```
