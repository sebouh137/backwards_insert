```python
#additional material not included in the dd4hep file
def add_mat(eta, phi):
    theta=get_theta(eta)
    X0s=0
    #circular flange, modeled as a solid cylinder
    z=-120
    r=np.tan(theta)*z
    
    rad_length_Al=8.897
    if r<7.612 and r>3.5:
        length=3.9624
        X0s+=length/rad_length_Al
        
    #now for the cam-shaped flange
    z=-269.72
    OR1=11.106/2
    IR1=8.2346/2
    x1=0
    
    
    OR2=2.8440
    IR2=3.1475/2
    x2=-0.025*z
    
    x=z*np.tan(theta)*np.cos(phi)
    y=z*np.tan(theta)*np.sin(phi)
    
    r1=np.hypot(x, y)
    r2=np.hypot(x-x2, y)
    
    if r1<OR1 and r1>IR1 or r2<OR2 and r2>IR2:
        length=2.54
        X0s+=length/rad_length_Al
    
    
    
    return X0s
```
